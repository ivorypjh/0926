```python
# 숨어있는 숫자의 덧셈

# 문자열이 매개변수로 주어지고 그 문자열은
# 영문과 숫자로 구성
# 그 안에서 숫자만 찾아서 더하는데
# 연속되는 숫자는 이어진 숫자로 판별하는 문제

def solution(my_string):
    answer = 0
    # 연속되는 숫자인 경우 활용할 변수
    cnt = 0
    # 연속되는 숫자를 반영하기 위한 리스트 생성
    temp_list = [0]
    
    # 문자열을 하나씩 순회
    for index in range(len(my_string)):
        # 해당 위치의 문자가 숫자인 경우 실행
        try:
            # 숫자 변수형으로 변환
            temp = int(my_string[index])
            # 연속되는 숫자인 경우 이전 값에 10을 곱하고
            # 현재 값을 더함
            temp_list[cnt] = temp_list[cnt] * 10 + temp
        # 해당 위치의 문자가 숫자가 아닌 경우 실행
        # 숫자를 저장하는 리스트의 인덱스(cnt) 를 증가시켜서
        # 별개의 숫자로 인식하도록
        except:
            cnt += 1
            # 위의 방식에서 곱하기 10을 적용하기 위해서
            # 문자인 경우 0을 추가해서 에러가 나지 않도록 조정
            temp_list.append(0)
            
    answer = sum(temp_list)
    
    return answer


# 다른 풀이
def solution(my_string):
    # 문자열을 하나씩 가져와서 숫자인지 우선 판별
    # 숫자라면 바로 더하고 숫자가 아닌 경우 공백을 더함
    # 마지막으로 공백을 기준으로 분할해서 숫자로 변환하면
    # 자릴수를 맞춘 숫자로 반환이 됨
    s = ''.join(i if i.isdigit() else ' ' for i in my_string)
    return sum(int(i) for i in s.split())
```

```python
# 다항식 더하기

# 1개 이상의 항으로 이루어진 다항식의 계산이
# 문자열로 주어질 때 그를 계산해서 결과를 출력하는 문제
def solution(polynomial):
    answer = ''
    num_list = [0, 0]
    
    target_list = polynomial.split(' + ')
    
    for item in target_list:
        if 'x' in item:
            if item == 'x':
                num_list[0] += 1
            else:
                num_list[0] += int(item[:-1])
        else:
            num_list[1] += int(item)
    if num_list[0] != 1:
        str1 = str(num_list[0]) + 'x'
    else:
        str1 = 'x'
        
    str2 = str(num_list[1])
    
    if str1 == '0x':
        answer = str2
    elif str2 == '0':
        answer = str1
    else:
        answer = str1 + ' + ' + str2
    
    return answer
```

```python
# 최댓값 만들기

# 숫자들을 가지고 있는 리스트가 주어졌을 때
# 그 숫자들로 만들 수 있는 최댓값을 반환하는 문제

def solution(numbers):
    
    maxi = [a * b for a in numbers for b in numbers if (a != b) \
            | (numbers.index(a) != numbers.index(b))]
    answer = max(maxi)
    
    return answer
```
